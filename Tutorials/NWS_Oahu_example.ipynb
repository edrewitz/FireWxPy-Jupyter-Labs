```python
'''
In this example we will make graphics of the National Weather Service Maximum/Minimum Temperature Forecast, Maximum/Minimum Relative Humidity Forecast 
for the island of Oahu. 

Since we are making 4 plots, we will download the data outside of the function and pass it in. 

The images will save to the following path structure f:Weather Data/NWS Forecasts/Type Of Forecast (i.e. NWS Max T)/HI/Island

If the forecast is for statewide like it will be in this example, the Island will be: STATE
Thus the full path to the graphics for the NWS Maximum Temperature Forecast will be f:Weather Data/NWS Forecasts/NWS Max T/HI/STATE

In this case, the island will be Oahu

'''
# Imports the needed modules

from firewxpy import nws_temperature_forecast_hawaii as t
from firewxpy import nws_relative_humidity_forecast_hawaii as rh
from firewxpy import NDFD_CONUS_Hawaii
```

```python
# Now we will download the data to pass into the functions
directory_name = '/SL.us008001/ST.opnl/DF.gr2/DC.ndfd/AR.hawaii/'

grbs_maxrh, ds_maxrh, count_short_maxrh, count_extended_maxrh = NDFD_CONUS_Hawaii.download_NDFD_grids(directory_name, 'ds.maxrh.bin')
grbs_minrh, ds_minrh, count_short_minrh, count_extended_minrh = NDFD_CONUS_Hawaii.download_NDFD_grids(directory_name, 'ds.minrh.bin')
grbs_maxt, ds_maxt, count_short_maxt, count_extended_maxt = NDFD_CONUS_Hawaii.download_NDFD_grids(directory_name, 'ds.maxt.bin')
grbs_mint, ds_mint, count_short_mint, count_extended_mint = NDFD_CONUS_Hawaii.download_NDFD_grids(directory_name, 'ds.mint.bin')
```

```python
# Now it is time to make our plots

t.plot_maximum_temperature_forecast(file_path='ds.maxt.bin', data_array=ds_maxt, count_short=count_short_maxt, count_extended=count_extended_maxt, island='oahu')
t.plot_minimum_temperature_forecast(file_path='ds.mint.bin', data_array=ds_mint, count_short=count_short_mint, count_extended=count_extended_mint, island='oahu')
rh.plot_maximum_relative_humidity_forecast(file_path='ds.maxrh.bin', data_array=ds_maxrh, count_short=count_short_maxrh, count_extended=count_extended_maxrh, island='oahu')
rh.plot_minimum_relative_humidity_forecast(file_path='ds.minrh.bin', data_array=ds_minrh, count_short=count_short_minrh, count_extended=count_extended_minrh, island='oahu')
```
