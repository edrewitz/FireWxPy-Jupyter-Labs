```python
'''
In this example we will make graphics of the National Weather Service Maximum/Minimum Temperature Forecast, Maximum/Minimum Relative Humidity Forecast 
for the CWA of NWS Juneau (AJK). 

Since we are making 4 plots, we will download the data outside of the function and pass it in. 

The images will save to the following path structure f:Weather Data/NWS Forecasts/Type Of Forecast (i.e. NWS Max T)/HI/CWA

If the forecast is for statewide like it will be in this example, the CWA will be: STATE
Thus the full path to the graphics for the NWS Maximum Temperature Forecast will be f:Weather Data/NWS Forecasts/NWS Max T/AK/AJK

In this case, the CWA will be AJK

'''
# Imports the needed modules

from firewxpy import nws_temperature_forecast_alaska as t
from firewxpy import nws_relative_humidity_forecast_alaska as rh
from firewxpy import NDFD_Alaska, NDFD_CONUS_Hawaii
```

```python
# Now we will download the data to pass into the functions
directory_name = '/SL.us008001/ST.opnl/DF.gr2/DC.ndfd/AR.alaska/'

ds_short_maxrh, ds_extended_maxrh = NDFD_Alaska.get_short_and_extended_grids('ds.maxrh.bin')
ds_short_minrh, ds_extended_minrh = NDFD_Alaska.get_short_and_extended_grids('ds.minrh.bin')
ds_short_maxt, ds_extended_maxt = NDFD_Alaska.get_short_and_extended_grids('ds.maxt.bin')
ds_short_mint, ds_extended_mint = NDFD_Alaska.get_short_and_extended_grids('ds.mint.bin')

grbs_maxrh, ds_maxrh, count_short_maxrh, count_extended_maxrh = NDFD_CONUS_Hawaii.download_NDFD_grids(directory_name, 'ds.maxrh.bin')
grbs_minrh, ds_minrh, count_short_minrh, count_extended_minrh = NDFD_CONUS_Hawaii.download_NDFD_grids(directory_name, 'ds.minrh.bin')
grbs_maxt, ds_maxt, count_short_maxt, count_extended_maxt = NDFD_CONUS_Hawaii.download_NDFD_grids(directory_name, 'ds.maxt.bin')
grbs_mint, ds_mint, count_short_mint, count_extended_mint = NDFD_CONUS_Hawaii.download_NDFD_grids(directory_name, 'ds.mint.bin')
```

```python
# Now it is time to make our plots

t.plot_maximum_temperature_forecast(file_path='ds.maxt.bin', ds_short=ds_short_maxt, ds_extended=ds_extended_maxt, ds=ds_maxt, count_short=count_short_maxt, count_extended=count_extended_maxt, cwa='ajk')
t.plot_minimum_temperature_forecast(file_path='ds.mint.bin', ds_short=ds_short_mint, ds_extended=ds_extended_mint, ds=ds_mint, count_short=count_short_mint, count_extended=count_extended_mint, cwa='ajk')
rh.plot_maximum_relative_humidity_forecast(file_path='ds.maxrh.bin', ds_short=ds_short_maxrh, ds_extended=ds_extended_maxrh, ds=ds_maxrh, count_short=count_short_maxrh, count_extended=count_extended_maxrh, cwa='ajk')
rh.plot_minimum_relative_humidity_forecast(file_path='ds.minrh.bin', ds_short=ds_short_minrh, ds_extended=ds_extended_minrh, ds=ds_minrh, count_short=count_short_minrh, count_extended=count_extended_minrh, cwa='ajk')
```
