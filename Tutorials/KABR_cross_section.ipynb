```python
'''
In this example we will use FireWxPy to recreate my favorite Dr. Brian Tang time cross-section graphics from the old U-Albany Model Guidance page for WxChallenge! 
The current WxChallenge city is Aberdeen, SD so we will use KABR as the station ID

In this example we will create GFS0p25 and NAM cross-sections showing the lower atmospheric winds and Temperature/RH/Vertical Velocity/Wind. 

'''
# Import needed packages
from firewxpy import time_cross_sections
from firewxpy import model_data
```

```python
'''
To minimize the amount of data requests we will use the method that we would use if we were to make many cross-sections at once for different locations
Eventhough we are only doing KABR in this example, we are creating 4 different graphics and there is no reason to ping the NCEP/NOMADS server more than 2
times since we are downloading the data of 2 different models. 

Since both locations are in CONUS, setting the area to CONUS is fine. We set all the manual lat/lon coordinate bounds to None since we are using a pre-defined area. 

'''
gfs = model_data.get_nomads_opendap_data('GFS0p25', 'conus', None, None, None, None)
nam = model_data.get_nomads_opendap_data('NAM', 'conus', None, None, None, None)
```

```python
'''
Now we make our graphics. Since we downloaded the data outside of the function and are passing it in, we need to change two of the optional arguments
1) data=False is the default ---> We need to set data=True to tell the function not to download the data inside of the function. 
2) ds=None is the default ---> We need to set ds to the dataset we want to pass in so for the GFS we will set ds=gfs and for the NAM set ds=nam

'''

time_cross_sections.plot_lower_atmosphere_wind('GFS0p25', 'KABR', 'KABR.png', data=True, ds=gfs)
time_cross_sections.plot_lower_atmosphere_wind('NAM', 'KABR', 'KABR.png', data=True, ds=nam)
time_cross_sections.plot_lower_atmosphere_temperature_rh_vertical_velocity_wind('GFS0p25', 'KABR', 'KABR.png', data=True, ds=gfs)
time_cross_sections.plot_lower_atmosphere_temperature_rh_vertical_velocity_wind('NAM', 'KABR', 'KABR.png', data=True, ds=nam)
```
