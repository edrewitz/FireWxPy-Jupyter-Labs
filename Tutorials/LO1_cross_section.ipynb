```python
'''
In this example we will use FireWxPy to create both a NAM and a RAP time cross-section forecast using the custom lat/lon of the LO1 buoy (43.621°N 77.401°W) on Lake Ontario
There is currently a LLAP-B Lake-Effect Snow event on Lake Ontario and my alma mater (SUNY Oswego) is experiencing Lake-Effect Snow. 

'''
# Imports the time cross section class from firewxpy
from firewxpy import time_cross_sections
```

```python
'''
Since we are analyzing the Lake-Effect Snow potential, the best graphic for that would be from the plot_lower_atmosphere_temperature_rh_vertical_velocity_wind() function as we will see
vertical velocity (shaded), temperature contours, RH contours, the dendridic growth zone depth (between -12°C and -18°C) and wind barbs. 

In this example you will see prime conditions for Lake-Effect Snow which are:

1) 13-15°C temperature difference between the surface and 850mb
2) Weak speed and directional wind shear ---> Straight lined flow at 30-45 MPH in this case
3) Instability (strong negative vertical velocity [blue shading on the graphic])

'''
time_cross_sections.plot_lower_atmosphere_temperature_rh_vertical_velocity_wind('NAM', 'custom', 'LO1.png', longitude=-77.401, latitude=43.621)
time_cross_sections.plot_lower_atmosphere_temperature_rh_vertical_velocity_wind('RAP', 'custom', 'LO1.png', longitude=-77.401, latitude=43.621)
```
